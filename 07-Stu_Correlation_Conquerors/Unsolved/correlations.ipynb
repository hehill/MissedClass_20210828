```python
# Dependencies
import pandas as pd
import sklearn.datasets as dta
import scipy.stats as st
import matplotlib.pyplot as plt
```

```python
# Read in the wine recognition data set from sklearn and load into Pandas
data = dta.load_wine()
wine_data = pd.DataFrame(data.data,columns=data.feature_names)
wine_data.head()
```

```python
# Plot malic_acid versus flavanoids on a scatterplot
plt.scatter(x = wine_data.malic_acid, y = wine_data.flavanoids)
plt.title("Malic Acid vs Flavanoid Content in Wines")
plt.xlabel("Malic Acid")
plt.ylabel("Flavanoids")
```

```python
# Calculate the correlation coefficient between malic_acid and flavanoids
ma = wine_data.malic_acid
fl = wine_data.flavanoids
correlation = st.pearsonr(ma,fl)
print(f"The correlation between both factors is {round(correlation[0],2)}")
```

```python
# Plot alcohol versus colour_intensity on a scatterplot
plt.scatter(x = wine_data.alcohol, y = wine_data.color_intensity)
plt.title("Alcohol vs Color Intensity of Wines")
plt.xlabel("Alcohol")
plt.ylabel("Color Intensity")
```

```python
# Calculate the correlation coefficient between alcohol and color_intensity
correlation = st.pearsonr(wine_data.alcohol,wine_data.color_intensity)
print(f"The correlation between both factors is {round(correlation[0],2)}")
```

```python
# BONUS: Generate the correlation matrix and find the strongest positive and negative correlations
corrMatrix = wine_data.corr()
corrMatrix
```

```python
import seaborn as sn
sn.heatmap(corrMatrix, annot = True)
plt.show()
```

```python
print(f"The highest correlation is between flavanoids and total phenol content ({round(corrMatrix.loc['flavanoids', 'total_phenols'],2)}). The second highest is between flavanoids and the OD280/OD315 ({round(corrMatrix.loc['flavanoids', 'od280/od315_of_diluted_wines'],2)}).")
```
