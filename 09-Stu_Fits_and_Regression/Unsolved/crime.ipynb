```python
# Dependencies
from matplotlib import pyplot as plt
from scipy import stats
from scipy.stats import linregress
import numpy as np
import pandas as pd
```

```python
# Load crime data set into pandas
crime_data = pd.read_csv("../Resources/crime_data.csv")
crime_data.head()
crime_data.keys()
```

```python
# Generate a scatter plot of violent crime rate versus year
plt.scatter(crime_data.year,crime_data["violent crime rate"])
plt.xlabel('Year')
plt.ylabel('Violent Crime Rate')
plt.show()
```

```python
# Perform a linear regression on violent crime rate versus year
x_values = crime_data.year
y_values = crime_data["violent crime rate"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

```

```python
# Create equation of line to calculate predicted violent crime rate
print(line_eq)
```

```python
# Plot the linear model on top of scatter plot 
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(2002,700),fontsize=15,color="red")
plt.xlabel('Year')
plt.ylabel('Violent Crime Rate')
```

```python
# Repeat plotting scatter and linear model for murder rate versus year
x_values = crime_data.year
y_values = crime_data[" murder and non-negligent manslaughter rate"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(2002,8),fontsize=15,color="red")
plt.xlabel('Year')
plt.ylabel('Murder rate')
```

```python
# Repeat plotting scatter and linear model for aggravated assault versus year
x_values = crime_data.year
y_values = crime_data["aggravated assault rate"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(2002,400),fontsize=15,color="red")
plt.xlabel('Year')
plt.ylabel('Aggravated Assault rate')
```
